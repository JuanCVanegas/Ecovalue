```python
import pandas as pd
```

```python
df = pd.read_csv('Denominador-1-modificado-4.csv')
df[ 'Dates' ] = pd.to_datetime(df['Dates'])
df.head()
```

```python
df['Revenue'] = df['Revenue'].interpolate('linear')
df['Cost of revenue'] = df['Cost of revenue'].interpolate('linear')
df['Operating expenses'] = df['Operating expenses'].interpolate('linear')
df['NOI'] = df['NOI'].interpolate('linear')
df['CAPEX'] = df['CAPEX'].interpolate('linear')
df['CWK'] = df['CWK'].interpolate('linear')
df['FCF'] = df['FCF'].interpolate('linear')
```
